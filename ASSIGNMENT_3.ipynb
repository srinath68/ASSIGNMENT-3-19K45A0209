Read mobile dataset using pandas, and preprocess it (fill null values).

```
import pandas as pd
s=pd.read_csv('/content/mobile_cleaned-1549119762886.csv')
print(s)
s.describe()
```

```
s.isnull()
```

```
s.isnull().sum()
```

```
s['weight']=s['weight'].fillna(s['weight'].mean())
s['primary_camera_resolution']=s['primary_camera_resolution'].fillna(s['primary_camera_resolution'].mean())
s['video_resolution']=s['video_resolution'].fillna(s['video_resolution'].mean())
s['talk_time']=s['talk_time'].fillna(s['talk_time'].mean())
s['no_of_reviews_in_gsmarena_in_week']=s['no_of_reviews_in_gsmarena_in_week'].fillna(s['no_of_reviews_in_gsmarena_in_week'].mean())
s['phone_height']=s['phone_height'].fillna(s['phone_height'].mean())
s['stand_by_time']=s['stand_by_time'].fillna(s['stand_by_time'].mean())
```

Compute outer product of 2 vectors. Eg:original matrix:[[1, 0], [0, 1]][[1, 2], [3, 4]]Outer product of the said two vectors:[[1 2 3 4] [0 0 0 0] [0 0 0 0] [1 2 3 4]]

```
import numpy as np
p = [[1, 0], [0, 1]]
q = [[1, 2], [3, 4]]
print("original matrix:")
print(p)
print(q)
result = np.outer(p, q)
print("Outer product of the said two vectors:")
print(result)
```

Write a NumPy program to compute the condition number of a given matrix. 5.Original matrix: 6.[[1 2]
[3 4]] 8.Condition number of the said matrix: 9.14.9330343737

```
import numpy as np
from numpy import linalg as LA
a = np.array([[1, 0, -1], [0, 1, 0], [1, 0, 1]])
print("Original matrix:")
print(a)
print("The condition number of the said matrix:")
print(LA.cond(a))
```

write a NumPy program to compute the sum of the diagonal element of a given array(matrix).

```
import numpy as np
m = np.arange(6).reshape(2,3)
print("Original matrix:")
print(m)
result =  np.trace(m)
print("Condition number of the said matrix:")
print(result)
```
